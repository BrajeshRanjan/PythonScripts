```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from mlxtend.plotting import plot_linear_regression
from mlxtend.plotting import plot_decision_regions
import pylab as pl
```

```python
%matplotlib notebook
```

```python
def f(size):
    '''
    Returns a sample with 'size' instances without noise.
    
    '''
    
    x = np.linspace(0, 4.5, size)
    y = 2 * np.sin(x * 1.5)
    return (x,y)

def sample(size):
    '''
    
    Returns a sample with 'size' instances.
    
    '''
    x = np.linspace(0, 4.5, size)
    y = 2 * np.sin(x * 1.5) + pl.randn(x.size)
    return (x,y)

#f_x, f_y = f(150)
#plt.plot(f_x, f_y, 'r')
x, y = sample(150)
plt.plot(x, y, 'b.')
```

```python
df = pd.DataFrame(list(zip(x,y)),columns = ['x', 'y'])
df.head()
```

```python
linear_model = LinearRegression()
linear_model.fit(df[['x']],df['y'])
predictions = linear_model.predict(df[['x']])
```

```python
plt.plot(df['x'],df['y'], 'b.')

#plot_linear_regression(x,y,linear_model, fit_style='r')

plt.plot(df['x'], predictions, color = 'r')
```

```python
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
model = KNeighborsRegressor(n_neighbors=21)
model.fit(df[['x']], df['y'])
predictions = model.predict(df[['x']])
```

```python
plt.scatter(df['x'], df['y'], c = 'b')
plt.plot(df[['x']], predictions, 'r')
```

```python
from sklearn.svm import SVR
```

```python
svr = SVR(kernel = 'rbf')
svr.fit(df[['x']],df['y'])
predictions = svr.predict(df[['x']])
```

```python
plt.scatter(df['x'], df['y'], c = 'b')
plt.plot(df[['x']], predictions, 'r')
```
