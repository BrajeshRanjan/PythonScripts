```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 12, 10
```

```python
np.random.seed(10)
x = np.array([i*np.pi/180 for i in range(60, 300, 4)])
y = np.sin(x) + np.random.normal(0, 0.15, len(x))
```

```python
data = pd.DataFrame(np.column_stack([x,y]),columns=['x_1','y'])
```

```python
plt.plot(data['x_1'],data['y'],'.')
```

```python
np.__version__ #getting numpy version
```

```python
#adding higher powers of x in the equatuion
for i in range(2,17):
    colname = 'x_' + str(i)
    data[colname] = data['x_1'] ** i
```

```python
data.head()
```

```python
x = data[['x_1']]
y = data['y']
```

```python
x.head()
```

```python
linear_regression = LinearRegression(normalize=True)
linear_regression.fit(x,y)
y_pred = linear_regression.predict(x)
```

```python
plt.plot(data['x_1'],y_pred)
plt.plot(data['x_1'],y, '.')
```

## Multiple Linear Regression (High Variance)

```python
X = data.drop(['y'], axis = 1)
y = data['y']
```

```python
X.head()
```

```python
linear_regression = LinearRegression()
linear_regression.fit(X,y)
y_pred = linear_regression.predict(X)
```

```python
plt.plot(data['x_1'],y_pred)
plt.plot(data['x_1'],y, '.')
```

```python
def linear_regression(power):
    columns = []
    for i in range (1, power+1):
        columns.append('x_' + str(i))
    X = data[columns]
    y = data['y']
    model = LinearRegression(normalize=True)
    model.fit(X,y)
    y_pred = model.predict(X)
    plt.plot(data['x_1'], y_pred)
    plt.plot(data['x_1'], data['y'], '.')        
```

```python
linear_regression(4)
```

## Regularization (Ridge)

```python
def ridge_regression(alpha):
    X = data.drop(['y'], axis = 1)
    y = data['y']
    ridge_regression = Ridge(alpha=alpha, normalize=True)
    ridge_regression.fit(X,y)
    y_pred = ridge_regression.predict(X)
    plt.plot(data['x_1'], y_pred)
    plt.plot(data['x_1'], data['y'], '.') 
```

```python
alpha_vals = [1e-15, 1e-10, 1e-5, 1e-2, 1e5, 1e10]
ridge_regression(0.01)
```
